```python
import json
import requests
response = requests.get("https://api.stackexchange.com/2.3/questions?order=desc&sort=activity&site=stackoverflow")
print(response)
```

```python
#now printing the files in json format

print(response.json())
```

```python
# to just print the items from the entire response

print(response.json()['items'])
```

```python
# iteratig through the items and printing the titles and corresponding links

for data in response.json()['items']:
    print(data['title'])
    print(data['link'])
```

```python
#printing the only titles and links where the answer count is zero or it unanswered

for data in response.json()['items']:
    if data['answer_count']==0:
        print(data['title'])
        print(data['link'])
    else:
        print("it is skipped")
print()
```
